```python
# Transform the dependent variable to the relative percentage rank across the universe at a specific time
df['m+1_returns'] = df['m+1']
df.loc[:,'m+1'] = df.groupby(['DATE'])['m+1'].transform(lambda x: x.rank(pct = True))
```

```python
# Classify the raw columns into several subclasses
identifier_cols = ['DATE','COMNAM','NCUSIP','PERMNO','PERMCO','SICCD','TICKER_PERMNO']
cross_sectional_cols=['accru', 'adv_sale', 'aftret_eq', 'aftret_equity','aftret_invcapx', 'at_tu', 'bm', 'CAPEI', 'capital_rat', 'cash_debt',
       'cash_lt', 'cash_rat', 'cfm', 'curr_debt', 'curr_rat', 'de_rat','debt_assets', 'debt_at', 'debt_capit', 'debt_ebitd', 'debt_invcap',
       'divyield', 'dltt_be', 'equity_invcap', 'evm', 'gpm', 'GProf', 'intcov','intcov_rat', 'invt_act', 'lt_debt', 'lt_ppent', 'npm', 'ocf_lct',
       'opmad', 'opmbd', 'pay_tu', 'pcf', 'pe_ex', 'pe_inc', 'pretret_earnat','pretret', 'profit_lct', 'ps', 'ptb', 'ptpm', 'quick_rat', 'rd_sale',
       'rect_act', 'rect_tu', 'roa', 'roce', 'roe', 'sale_equity','sale_invcap', 'short_debt', 'staff_sale', 'totdebt_invcap']
sector_cols = ['Agriculture, Forestry and Fishing', 'Construction','Finance, Insurance', 'Manufacturing', 'Mining', 'Nonclassifiable',
       'Public Administration', 'Real Estate', 'Retail Trade', 'Services','Transportation, Communications, Electric, Gas and Sanitary service',
       'Wholesale Trade', 'not used']
time_series_cols= ['m-12','m-11', 'm-10', 'm-9', 'm-8', 'm-7', 'm-6', 'm-5', 'm-4', 'm-3', 'm-2','m-1']
all_cols = cross_sectional_cols + sector_cols + time_series_cols
Y = ['m+1']
```

## Step 3: Train the Autoencoder

```python
# Import necessary built-in libraries
import sklearn as sk
import sklearn.metrics
import sklearn.utils 
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, Concatenate, Input, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint,ReduceLROnPlateau
```

```python
# Specify the training data sample 

start_date='1971-01-31'
end_date='1981-01-31'
test_end_date='1981-02-28'

mask_train=(df[['DATE']]>=start_date) & (df[['DATE']]<end_date)
mask_test=(df[['DATE']]>=end_date) & (df[['DATE']]<test_end_date)


fin_cols=cross_sectional_cols + sector_cols # cross-sectional inputs does not include ts data
x = [df[fin_cols].values,df[time_series_cols].values] # WAS df[all_cols]
y = df[Y].values[:,0]


x_train, x_test, y_train, y_test=[i[mask_train['DATE'],:] for i in x], [i[mask_test['DATE'],:] for i in x], y[mask_train['DATE']], y[mask_test['DATE']]

x_tr=np.array(x_train[0])
x_te=np.array(x_test[0])
```

### Autoencoder

```python
# Import Autoencoder class from the user defined python file
from utils.Autoencoder import Autoencoder
```

```python
len(cross_sectional_cols) # Orignial input dimension excluding 13 sector feature
```

```python
x_train_auto=x_tr[:,:58] # Only use the financial raio features excluding 13 sector identifications
x_test_auto=x_te[:,:58]
x_train_auto.shape # Training data shape
```

```python
# Train autoencoder without the sector factors by using 10 year datapoints 1971-1981
# 58 sectors in total
# Use the left cross_sectional_cols to walk through the autoencoder

hidden_units=30  #<--- User defined latent representation/Encoded dimension of the original data


# Fit the training data
ae = Autoencoder(ENCODED_DIM=hidden_units, sparsity=0.5, regu=0.001, learning_rate=0.01, batch_size=256, epochs = 400)
ae.fit(x_train_auto)
```

## Step 4: Transform all the data into the trained autoencoder

```python
all_fin_x=df[fin_cols].values # 1971-2017 all the data points 

fin_non_sector=all_fin_x[:,:58] # cross sectional cols of dimension 58 
fin_sector=all_fin_x[:,58:]     # sector dummies of dimension 13
```

```python
fin_non_sector.shape
```

```python
# Transform all the non sector indicators into lower dimension using the trained autoecoder 
transformed_financial_ratio = ae.transform(fin_non_sector)

# Concate the sector features to the transformed cross section features
summarized_data_fc= np.hstack((transformed_financial_ratio,fin_sector))
```
