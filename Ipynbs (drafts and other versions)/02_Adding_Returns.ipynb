```python
dates = pd.date_range('1971-01-31','2019-12-31', freq='M').tolist()
```

```python
cols = ['m-12','m-11','m-10','m-9','m-8','m-7','m-6','m-5','m-4','m-3','m-2','m-1','m+1']
```

```python
new_return_table=pd.DataFrame()
for d in dates:
    print(d)
    time_slice = list(pd.date_range(end=d,periods=12, freq='M').tolist() + pd.date_range(start=((d+datetime.timedelta(days=1))),periods=1, freq='M').tolist())
    new_temp_df = return_table.loc[:,time_slice]
    new_temp_df.columns=cols
    new_temp_df=new_temp_df.dropna(how='all')
    new_temp_df['cusip'] = new_temp_df.index 
    new_temp_df['DATE'] = d 
    new_return_table = new_return_table.append(new_temp_df,ignore_index=True)
```
