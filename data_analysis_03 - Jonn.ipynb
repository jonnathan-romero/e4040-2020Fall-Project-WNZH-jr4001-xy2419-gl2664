```python
df = df.loc[df.isna().sum(axis=1)<20,:]
```

```python
list(zip(list(df.columns),list(df.count()/df.shape[0])))
```

```python
df=df[~df['m+1'].isna()]
```

```python
list(zip(list(df.columns),list(df.count()/df.shape[0])))
```

```python
inverse_cols=['evm','pcf', 'pe_ex', 'pe_inc','ps','ptb'] 
```

```python
for ele in inverse_cols:
    df[[ele]]=df[[ele]].apply(lambda x: 1/x)
```

```python
import numpy as np
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
list(zip(list(df.columns),list(df.count()/df.shape[0])))
```

```python
numeric_cols=['accru', 'adv_sale', 'aftret_eq', 'aftret_equity',
       'aftret_invcapx', 'at_tu', 'bm', 'CAPEI', 'capital_rat', 'cash_debt',
       'cash_lt', 'cash_rat', 'cfm', 'curr_debt', 'curr_rat', 'de_rat',
       'debt_assets', 'debt_at', 'debt_capit', 'debt_ebitd', 'debt_invcap',
       'divyield', 'dltt_be', 'equity_invcap', 'evm', 'gpm', 'GProf', 'intcov',
       'intcov_rat', 'invt_act', 'lt_debt', 'lt_ppent', 'npm', 'ocf_lct',
       'opmad', 'opmbd', 'pay_tu', 'pcf', 'pe_ex', 'pe_inc', 'pretret_earnat',
       'pretret', 'profit_lct', 'ps', 'ptb', 'ptpm', 'quick_rat', 'rd_sale',
       'rect_act', 'rect_tu', 'roa', 'roce', 'roe', 'sale_equity',
       'sale_invcap', 'short_debt', 'staff_sale', 'totdebt_invcap', 'm-12',
       'm-11', 'm-10', 'm-9', 'm-8', 'm-7', 'm-6', 'm-5', 'm-4', 'm-3', 'm-2',
       'm-1', 'm+1']
```

```python
for col in numeric_cols:
    df[col] = np.where(df[col].isna(),df.groupby(['DATE'])[col].transform('mean'),df[col])
```

```python
list(zip(list(df.columns),list(df.count()/df.shape[0])))
```

```python
from scipy.stats import zscore
for col in numeric_cols:
    df[col] = df.groupby(['DATE'])[col].transform(lambda x : zscore(x,ddof=1))
```

```python
df.to_csv('complete_universe_cleaned_normalized_jonn.csv')
```
